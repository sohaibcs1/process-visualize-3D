```python
import bioformats
import javabridge



file_path = '1.czi'

# Get metadata
metadata = bioformats.get_omexml_metadata(file_path)
root = bioformats.OMEXML(metadata)
pixels = root.image().Pixels

# Collect channel names and colors
channel_info = []
for i in range(pixels.SizeC):
    channel = pixels.Channel(i)
    name = channel.Name
    try:
        # Color values might be stored as an RGB triplet (hex or individual components)
        color = channel.Color  # This returns the color as an integer (if available)
        # Extract RGB components from the color integer (if applicable)
        r = (color >> 16) & 0xFF
        g = (color >> 8) & 0xFF
        b = color & 0xFF
        channel_info.append((name, f'RGB({r}, {g}, {b})'))
    except AttributeError:
        channel_info.append((name, "No color information available"))

# Display channel names and their colors
print("Channel Information:")
for i, (name, color) in enumerate(channel_info):
    print(f"Channel {i + 1}: {name}, Color: {color}")


```
