```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu

# Load the dataset
file_path = '5 (2).xlsx'
data = pd.read_excel(file_path)

# Step 1: Extract 'Cell Line' and 'Day' from the 'Cell_Lines' column
data['Cell Line'] = data['Cell_Lines'].apply(lambda x: x.split(' ')[0])  # Extract cell line
data['Day'] = data['Cell_Lines'].apply(lambda x: x.split(' ')[-1])       # Extract day

# Define features and grouping variable
features = data.columns[1:-2]  # Numerical features
group = "Cell Line"            # Grouping variable
unique_groups = data[group].unique()

# Step 2: Create violin plots with improved aesthetics and Mann-Whitney U test
for feature in features:
    plt.figure(figsize=(12, 6))
    
    # Create the violin plot
    ax = sns.violinplot(x=group, y=feature, data=data, palette="muted", inner="point")
    plt.title(f"Violin Plot with Significance for {feature}", fontsize=16)
    plt.ylabel(feature, fontsize=12)
    plt.xlabel(group, fontsize=12)
    
    # Get x-coordinates of each group
    x_coords = [tick for tick in range(len(unique_groups))]

    # Perform pairwise Mann-Whitney U tests for each unique pair of groups
    pairs = [(unique_groups[i], unique_groups[j]) for i in range(len(unique_groups)) for j in range(i + 1, len(unique_groups))]
    y_max = data[feature].max()
    y_min = data[feature].min()
    offset = 0.05 * (y_max - y_min)  # Adjust for visual spacing
    line_spacing = offset * 2.5  # Additional space between lines

    for pair_idx, (g1, g2) in enumerate(pairs):
        group1_data = data[data[group] == g1][feature].dropna()
        group2_data = data[data[group] == g2][feature].dropna()
        stat, p_value = mannwhitneyu(group1_data, group2_data, alternative='two-sided')

        # Add significance stars based on p-value
        if p_value < 0.001:
            stars = "***"
        elif p_value < 0.01:
            stars = "**"
        elif p_value < 0.05:
            stars = "*"
        else:
            continue  # Skip not significant results

        # Align lines above violins
        x1 = unique_groups.tolist().index(g1)
        x2 = unique_groups.tolist().index(g2)
        y = y_max + offset + (pair_idx * line_spacing)  # Add spacing for each line
        plt.plot([x1, x2], [y, y], color='black', linewidth=1.5)
        plt.text((x1 + x2) * 0.5, y + offset / 2, stars, ha='center', va='bottom', color='black', fontsize=12)
    
    # Rotate x-axis labels for clarity
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Extract cell lines and days from the 'Cell_Lines' column
data['Cell Line'] = data['Cell_Lines'].apply(lambda x: x.split(' ')[0])  # Extract cell line (e.g., MCF10A)
data['Day'] = data['Cell_Lines'].apply(lambda x: x.split(' ')[-1])  # Extract day (e.g., Day 5)

# Map cell lines and days to distinct colors and markers
cell_line_colors = {
    'MCF10A': 'blue',
    'MCF7': 'green',
    'MDA-MB-231': 'red',
    'MDA-MB-468': 'purple'
}
day_markers = {
    '5': 'o',
    '7': 's'
}
color_map = data['Cell Line'].map(cell_line_colors)
marker_map = data['Day'].map(day_markers)

# Standardize the numeric data for PCA and t-SNE
numeric_data = data.select_dtypes(include=['float64', 'int64'])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# Perform PCA with 3 components
pca = PCA(n_components=3)
pca_result = pca.fit_transform(scaled_data)

# Perform t-SNE with 3 components and optimized parameters
tsne = TSNE(n_components=3, random_state=42, perplexity=5, learning_rate=100, n_iter=250)
tsne_result = tsne.fit_transform(scaled_data)

# Plot 3D PCA
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
for line in cell_line_colors.keys():
    for day in day_markers.keys():
        mask = (data['Cell Line'] == line) & (data['Day'] == day)
        ax.scatter(
            pca_result[mask, 0],
            pca_result[mask, 1],
            pca_result[mask, 2],
            c=cell_line_colors[line],
            marker=day_markers[day],
            label=f'{line} Day {day}',
            alpha=0.8
        )
ax.set_title('3D PCA: Cell Line and Day')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.legend(loc='best', fontsize=8)
plt.tight_layout()
plt.show()

# Plot 3D t-SNE
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
for line in cell_line_colors.keys():
    for day in day_markers.keys():
        mask = (data['Cell Line'] == line) & (data['Day'] == day)
        ax.scatter(
            tsne_result[mask, 0],
            tsne_result[mask, 1],
            tsne_result[mask, 2],
            c=cell_line_colors[line],
            marker=day_markers[day],
            label=f'{line} Day {day}',
            alpha=0.8
        )
ax.set_title('3D t-SNE: Cell Line and Day')
ax.set_xlabel('t-SNE Component 1')
ax.set_ylabel('t-SNE Component 2')
ax.set_zlabel('t-SNE Component 3')
ax.legend(loc='best', fontsize=8)
plt.tight_layout()
plt.show()
```
